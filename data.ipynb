```python
def merge_csv_files(input_directory, output_directory):
    # Get list of all files in the input directory
    files = os.listdir(input_directory)
    print(f"Files in directory: {files}")

    # Create output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Create a dictionary to map district names with their coordinates
    district_files = {}
    for file in files:
        if file.endswith(".csv") and not file.startswith('.'):
            # Extract district name and coordinates
            if '_2024' in file:
                district_name = file.replace('_2024.csv', '')
            else:
                district_name = file.replace('.csv', '')
            base_name = '_'.join(district_name.split('_')[:-2])  # Get the district name without coordinates
            if base_name not in district_files:
                district_files[base_name] = []
            district_files[base_name].append(file)

    for district, district_files_list in district_files.items():
        # Separate the filenames into those for 2023 and 2024
        files_2023 = [f for f in district_files_list if not '_2024' in f]
        files_2024 = [f for f in district_files_list if '_2024' in f]
        
        for file_2023 in files_2023:
            # Find the matching 2024 file based on coordinates
            coords = '_'.join(file_2023.split('_')[-2:]).replace('.csv', '')
            file_2024 = f"{district}_2024_{coords}.csv"

            print(f"Checking for files: {file_2023} and {file_2024}")

            # Check if both files exist
            file_2023_path = os.path.join(input_directory, file_2023)
            file_2024_path = os.path.join(input_directory, file_2024)
            
            if os.path.exists(file_2023_path) and os.path.exists(file_2024_path):
                print(f"Merging files: {file_2023_path} and {file_2024_path}")
                
                # Read both CSV files
                df_2023 = pd.read_csv(file_2023_path)
                df_2024 = pd.read_csv(file_2024_path)
                
                # Concatenate the dataframes
                merged_df = pd.concat([df_2023, df_2024])

                # Remove duplicates, if any
                merged_df = merged_df.drop_duplicates()

                # Save the merged dataframe in the output directory
                merged_file_name = f"{district}_{coords}.csv"
                merged_file_path = os.path.join(output_directory, merged_file_name)
                merged_df.to_csv(merged_file_path, index=False)
                print(f'Merged {file_2023} and {file_2024} into {merged_file_path}')
            else:
                if not os.path.exists(file_2023_path):
                    print(f"File missing: {file_2023_path}")
                if not os.path.exists(file_2024_path):
                    print(f"File missing: {file_2024_path}")
                print(f'Files for {district} with coordinates {coords} do not need merging or are missing.')
```

```python
input_directory_path = "/media/shuryanshgupta001/New Volume/Web Scraping/dataset/westbengal"  # Update this path to your input directory
output_directory_path = "/media/shuryanshgupta001/New Volume/Web Scraping/dataset/newwestbengal"  # Update this path to your output directory
merge_csv_files(input_directory_path, output_directory_path)

input_directory_path = "/media/shuryanshgupta001/New Volume/Web Scraping/dataset/bihar"  # Update this path to your input directory
output_directory_path = "/media/shuryanshgupta001/New Volume/Web Scraping/dataset/newbihar"  # Update this path to your output directory
merge_csv_files(input_directory_path, output_directory_path)
```
